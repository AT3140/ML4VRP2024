```python
import os

def convert_filenames_to_uppercase(folder_path):
    # Get list of files in the folder
    files = os.listdir(folder_path)
    
    # Iterate through each file
    for file in files:
        # Split the filename and extension
        name, ext = os.path.splitext(file)
        
        # Convert the filename to uppercase
        new_name = name.upper() + ext
        
        # Construct the paths for the original and new filenames
        old_path = os.path.join(folder_path, file)
        new_path = os.path.join(folder_path, new_name)
        
        # Rename the file
        os.rename(old_path, new_path)
        print(f"Renamed '{file}' to '{new_name}'")
```

```python
# Replace 'folder_path' with the path to your folder
folder_path = 'Instances/CVRP/json'
convert_filenames_to_uppercase(folder_path)
```

```python
import os
import pandas as pd

# Specify the folder path
#folder_path = './Instances/CVRP/vrp'
folder_path = './Instances/CVRPTW/txt'

# Get the list of file names in the folder
file_names = os.listdir(folder_path)
file_names_without_extension = [os.path.splitext(file)[0] for file in file_names]

# Create a pandas DataFrame with the file names as a column
#df_cvrp = pd.DataFrame(file_names_without_extension, columns=['Instances'])
df_cvrptw = pd.DataFrame(file_names_without_extension, columns=['Instances'])

# Display the DataFrame
#df_cvrp
df_cvrptw
```

```python
# Remove rows where 'Instances' column has ".DS_Store"
#df_cvrp = df_cvrp[df_cvrp['Instances'] != '.DS_Store']
df_cvrptw = df_cvrptw[df_cvrptw['Instances'] != '.DS_Store']
```

```python
df_cvrptw['Instances']
```

```python
# Assuming df is your DataFrame with 'Name' column containing [insName]

# Function to extract and round total distance from output
def extract_and_round_distance(output):
    nv_match = 0
    match = re.search(r'Number of vehicles:\s+(\d+)', output)
    if match:
        # Extract the number of vehicles from the match
        nv_match = int(match.group(1))
        
    distance_match = re.search(r"Total distance:\s+(\d+\.\d+)", output)
    if distance_match:
        total_distance = float(distance_match.group(1))
    else:
        total_distance = 'None'
    
    return nv_match, total_distance
```

```python
# Iterate over each name in the DataFrame
nv_values = []
evaluated_values = []
#for ins_name in df_cvrp['Instances']:
for ins_name in df_cvrptw['Instances']:
    # Construct the command
    #command = f"python evaluator.py cvrp {ins_name} ./../../Desktop/solutions-cvrp-to-test/{ins_name}.txt"
    command = f"python evaluator.py cvrptw {ins_name} ./../../Desktop/solutions-cvrptw-to-test/{ins_name}.txt"
    
    # Execute the command
    try:
        result = subprocess.run(command, shell=True, check=True, capture_output=True, text=True)
        # Extract and round total distance
        nv, total_distance = extract_and_round_distance(result.stdout)
        nv_values.append(nv)
        evaluated_values.append(total_distance)
    except subprocess.CalledProcessError as e:
        print(f"Error occurred while running the command for {ins_name}: {e}")
        nv_values.append(None)
        evaluated_values.append(None)
```

```python
# Add evaluated values to the DataFrame
df_cvrptw['evaluated-value'] = evaluated_values
df_cvrptw['nv-value'] = nv_values

# Display the DataFrame
df_cvrptw
```

```python
directory = './../../Desktop/Vrp-Set-HG/Vrp-Set-HG/'
```

```python
# Lists to store names and costs
names = []
costs = []

# Iterate through each file
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        # Extract name from the filename
        name = os.path.splitext(filename)[0]
        names.append(name)
        
        # Extract cost from the last line of the file
        with open(os.path.join(directory, filename), 'r') as file:
            last_line = file.readlines()[-1]
            cost = float(last_line.split()[1])  # Assuming cost is always the second word
            costs.append(cost)

# Create a DataFrame
data = {'Name': names, 'Cost': costs}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
#excel_filename = 'cvrp-bks.xlsx'
excel_filename = directory+'cvrptw-bks.xlsx'

df.to_excel(excel_filename, index=False)

print("DataFrame saved to Excel successfully.")
```

```python
import os

# Directory containing the files
#directory = 'solutions-cvrp-to-test/'

# Iterate through each file
for filename in os.listdir(directory):
    if filename.endswith(".txt"):
        file_path = os.path.join(directory, filename)
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # Remove the last line
        lines = lines[:-1]
        
        # Write the modified content back to the file
        with open(file_path, 'w') as file:
            file.writelines(lines)

print("Last line removed from each .txt file successfully.")
```
